```python
import numpy as np
import  matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('C:\\Users\\dell\\Desktop\\AI Workshop\\Ninapro_DB1.csv')
```

```python
drop_columns = ['stimulus','restimulus','repetition','rerepetition','subject']
dataset=dataset.drop(columns = drop_columns)
```

```python
column_to_plot = 'exercise'
plt.figure(figsize=(10, 5))
plt.hist(dataset[column_to_plot], bins=30, color='pink', edgecolor='black')
plt.xlabel(column_to_plot)
plt.ylabel('Frequency')
plt.title(f'Histogram for Exercise')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
