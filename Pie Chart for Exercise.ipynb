```python
import numpy as np
import  matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('C:\\Users\\dell\\Desktop\\AI Workshop\\Ninapro_DB1.csv')
```

```python
drop_columns = ['stimulus','restimulus','repetition','rerepetition','subject']
dataset=dataset.drop(columns = drop_columns)
```

```python
column_to_plot = 'exercise'
value_counts = dataset[column_to_plot].value_counts()
plt.figure(figsize=(10, 5))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Pie Chart for Exercise')
plt.show()
```
